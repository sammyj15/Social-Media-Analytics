```python
# Create new column to measure cross genre reach between the two artists; for every genre they have in common, they get a point
reach_level=[]
for i,row in df.iterrows():
    count=0
    for agenre in row['A Genres']:
        for bgenre in row['B Genres']:
            if agenre == bgenre:
                count += 1
            else:
                pass
    reach_level.append(count)
```

```python
df["In Genre Measure"] = reach_level 
```

```python
# new column to normalize in genre measure based on how many genres the artists belong to
ign = []
for i, row in df.iterrows():
    try:
        dd = row['In Genre Measure'] / (len(row['A Genres'])+len(row['B Genres']))
        ign.append(dd)
    except ZeroDivisionError as err:
        dd=0
        ign.append(dd)
df["In Genre Normalized"] = ign
```
